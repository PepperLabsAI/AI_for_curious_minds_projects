```python
# ============================================
# Intent Classification Engine
# Keyword + Scoring Method
# Beginner Friendly Version
# Ready to Run in Google Colab
# ============================================

print("🤖 Intent Classification Engine Started!")
print("Type 'exit' to stop.\n")

# -----------------------------------------
# Step 1: Define Intents and Keywords
# -----------------------------------------

# Each intent has a list of keywords
intents = {
    "greeting": ["hello", "hi", "hey", "good morning", "good evening"],
    "goodbye": ["bye", "goodbye", "see you", "exit"],
    "weather": ["weather", "temperature", "rain", "hot", "cold"],
    "name_query": ["your name", "who are you"],
    "thanks": ["thank you", "thanks"]
}

# -----------------------------------------
# Step 2: Function to Classify Intent
# -----------------------------------------

def classify_intent(user_input):
    user_input = user_input.lower()

    scores = {}  # Dictionary to store score for each intent

    # Check each intent
    for intent, keywords in intents.items():
        score = 0

        # Check each keyword in that intent
        for word in keywords:
            if word in user_input:
                score += 1

        scores[intent] = score

    # Find intent with highest score
    best_intent = max(scores, key=scores.get)

    # If highest score is 0 → Unknown intent
    if scores[best_intent] == 0:
        return "unknown"
    else:
        return best_intent


# -----------------------------------------
# Step 3: Generate Response
# -----------------------------------------

def get_response(intent):

    if intent == "greeting":
        return "Hello! How can I help you?"

    elif intent == "goodbye":
        return "Goodbye! Have a nice day!"

    elif intent == "weather":
        return "I cannot check live weather yet, but I hope it is pleasant outside!"

    elif intent == "name_query":
        return "I am a simple Intent Classification Bot."

    elif intent == "thanks":
        return "You are welcome!"

    else:
        return "Sorry, I did not understand that."


# -----------------------------------------
# Step 4: Chat Loop
# -----------------------------------------

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Goodbye!")
        break

    intent = classify_intent(user_input)
    response = get_response(intent)

    print("Detected Intent:", intent)
    print("Bot:", response)
    print()
```
