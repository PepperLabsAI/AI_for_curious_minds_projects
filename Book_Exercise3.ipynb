📌 Problem Statement
🚗 Predicting Fuel Efficiency of a Delivery Vehicle

A logistics company wants to predict the fuel consumption (liters per 100 km) of a delivery vehicle based on:

Distance traveled (in kilometers)

As distance increases, fuel consumption increases linearly.

We will build a Linear Regression model from scratch (without sklearn) to learn this relationship.

```python
# ============================================
# Linear Regression from Scratch
# Fuel Consumption Prediction
# Beginner Friendly
# Ready for Google Colab
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------
# Step 1: Create Dataset
# -----------------------------------------

# Distance in kilometers
X = np.array([10, 20, 30, 40, 50, 60, 70, 80])

# Fuel consumption in liters per 100 km
# (Slightly noisy linear data)
Y = np.array([2.1, 4.1, 6.2, 8.1, 10.2, 12.1, 14.2, 16.1])

# -----------------------------------------
# Step 2: Initialize Parameters
# -----------------------------------------

m = 0   # slope
b = 0   # intercept

learning_rate = 0.0005
epochs = 1000

n = len(X)

# -----------------------------------------
# Step 3: Gradient Descent
# -----------------------------------------

for i in range(epochs):

    Y_pred = m * X + b

    # Calculate gradients
    dm = (-2/n) * sum(X * (Y - Y_pred))
    db = (-2/n) * sum(Y - Y_pred)

    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

# -----------------------------------------
# Step 4: Print Learned Parameters
# -----------------------------------------

print("Training Complete!")
print("Learned slope (m):", round(m, 4))
print("Learned intercept (b):", round(b, 4))

# -----------------------------------------
# Step 5: Make Prediction
# -----------------------------------------

new_distance = 90
predicted_fuel = m * new_distance + b

print("\nFor distance =", new_distance, "km")
print("Predicted Fuel Consumption =", round(predicted_fuel, 2), "liters")

# -----------------------------------------
# Step 6: Visualization
# -----------------------------------------

plt.scatter(X, Y)
plt.plot(X, m * X + b)
plt.xlabel("Distance (km)")
plt.ylabel("Fuel Consumption (liters)")
plt.title("Linear Regression - Fuel Prediction")
plt.show()
```
