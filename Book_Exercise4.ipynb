📌 Problem Statement
🏦 Loan Approval Prediction System

A small finance company wants to automate loan approval decisions.

They approve a loan based on:

Applicant Age

Monthly Income (in thousands)

Credit Score

Existing Loan (0 = No, 1 = Yes)

The target variable is:

Loan Approved (1 = Yes, 0 = No)

Your task:

Build a Decision Tree Classifier that learns from past data and predicts whether a new applicant should get loan approval.

```python
# ============================================
# Decision Tree Classification Exercise
# Loan Approval Prediction
# Beginner Friendly
# Ready for Google Colab
# ============================================

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# -----------------------------------------
# Step 1: Create Dataset
# -----------------------------------------

data = {
    "Age": [25, 35, 45, 20, 50, 23, 40, 60, 48, 33],
    "Income": [25, 40, 60, 15, 80, 18, 55, 90, 70, 35],
    "CreditScore": [600, 650, 700, 580, 750, 590, 720, 800, 710, 640],
    "ExistingLoan": [0, 1, 0, 0, 1, 0, 1, 0, 1, 0],
    "LoanApproved": [0, 1, 1, 0, 1, 0, 1, 1, 1, 0]
}

df = pd.DataFrame(data)

# -----------------------------------------
# Step 2: Separate Features and Target
# -----------------------------------------

X = df[["Age", "Income", "CreditScore", "ExistingLoan"]]
y = df["LoanApproved"]

# -----------------------------------------
# Step 3: Create Decision Tree Model
# -----------------------------------------

model = DecisionTreeClassifier()
model.fit(X, y)

print("Model Training Complete!\n")

# -----------------------------------------
# Step 4: Make Prediction
# -----------------------------------------

# New applicant
new_applicant = [[30, 50, 690, 0]]

prediction = model.predict(new_applicant)

if prediction[0] == 1:
    print("Loan Approved ✅")
else:
    print("Loan Rejected ❌")

# -----------------------------------------
# Step 5: Visualize Decision Tree
# -----------------------------------------

plt.figure(figsize=(12,8))
plot_tree(model,
          feature_names=X.columns,
          class_names=["Rejected", "Approved"],
          filled=True)

plt.title("Decision Tree - Loan Approval")
plt.show()
```
