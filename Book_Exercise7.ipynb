🧊 Problem Statement

An agent is placed on a frozen lake.

Goal:

👉 Move from Start (S)
👉 Reach the Goal (G)
👉 Avoid holes (H)
👉 Move on safe tiles (F)

The environment is:

Grid world

Rewards:

+1 for reaching goal

0 otherwise

Your task:

Train an agent using Q-learning to reach the goal.

```python
# ============================================
# Simple Reinforcement Learning
# Q-Learning with Frozen Lake
# Beginner Friendly
# Ready for Google Colab
# ============================================

# Install library (only first time in Colab)
!pip install gymnasium

import gymnasium as gym
import numpy as np
import random

# -----------------------------------------
# Step 1: Create Environment
# -----------------------------------------

env = gym.make("FrozenLake-v1", is_slippery=False)

state_size = env.observation_space.n
action_size = env.action_space.n

print("Number of States:", state_size)
print("Number of Actions:", action_size)

# -----------------------------------------
# Step 2: Initialize Q-table
# -----------------------------------------

Q = np.zeros((state_size, action_size))

# Hyperparameters
learning_rate = 0.8
discount_rate = 0.95
epsilon = 1.0
epsilon_decay = 0.995
min_epsilon = 0.01
episodes = 1000

# -----------------------------------------
# Step 3: Q-Learning Training
# -----------------------------------------

for episode in range(episodes):

    state, _ = env.reset()
    done = False

    while not done:

        # Exploration vs Exploitation
        if random.uniform(0,1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[state])

        new_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        # Q-learning formula
        Q[state, action] = Q[state, action] + learning_rate * (
            reward + discount_rate * np.max(Q[new_state]) - Q[state, action]
        )

        state = new_state

    # Reduce exploration over time
    epsilon = max(min_epsilon, epsilon * epsilon_decay)

print("Training Complete!\n")

# -----------------------------------------
# Step 4: Test the Trained Agent
# -----------------------------------------

state, _ = env.reset()
done = False

print("Testing trained agent...\n")

while not done:
    action = np.argmax(Q[state])
    state, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    env.render()

if reward == 1:
    print("Agent reached the goal! ✅")
else:
    print("Agent failed ❌")
```
