📌 Problem Statement

A company wants to automatically detect emotions in customer feedback messages.

Given a text message, the system should classify it into one of the emotions:

joy

sadness

anger

fear

love

surprise

Your task:

Build a simple Emotion Detection system using NLP techniques.

```python
# ============================================
# Emotion Detection from Text
# Simple NLP Classification
# Beginner Friendly
# Ready for Google Colab
# ============================================

# Install dataset library
!pip install datasets

import pandas as pd
from datasets import load_dataset
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# -----------------------------------------
# Step 1: Load Emotion Dataset
# -----------------------------------------

dataset = load_dataset("emotion")

# Convert to pandas
train_df = pd.DataFrame(dataset["train"])
test_df = pd.DataFrame(dataset["test"])

print("Sample Data:")
print(train_df.head())

# -----------------------------------------
# Step 2: Vectorize Text (TF-IDF)
# -----------------------------------------

vectorizer = TfidfVectorizer(max_features=5000)

X_train = vectorizer.fit_transform(train_df["text"])
X_test = vectorizer.transform(test_df["text"])

y_train = train_df["label"]
y_test = test_df["label"]

# -----------------------------------------
# Step 3: Train Model
# -----------------------------------------

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

print("\nModel Training Complete!")

# -----------------------------------------
# Step 4: Evaluate Model
# -----------------------------------------

predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", round(accuracy * 100, 2), "%")

# -----------------------------------------
# Step 5: Emotion Prediction Function
# -----------------------------------------

emotion_labels = {
    0: "sadness",
    1: "joy",
    2: "love",
    3: "anger",
    4: "fear",
    5: "surprise"
}

def predict_emotion(text):
    vector = vectorizer.transform([text])
    prediction = model.predict(vector)[0]
    return emotion_labels[prediction]

# -----------------------------------------
# Step 6: Test with Custom Input
# -----------------------------------------

print("\nTry your own sentence.")
user_input = input("Enter text: ")

emotion = predict_emotion(user_input)
print("Predicted Emotion:", emotion)
```
