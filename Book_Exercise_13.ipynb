```python
# ============================================
# Text Similarity Checker using NLP
# Beginner Friendly
# Ready for Google Colab
# ============================================

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------------------
# Step 1: Define Two Text Documents
# -----------------------------------------

text1 = "Machine learning is a branch of artificial intelligence"
text2 = "Artificial intelligence includes machine learning techniques"

# -----------------------------------------
# Step 2: Convert Text to TF-IDF Vectors
# -----------------------------------------

vectorizer = TfidfVectorizer()

# Combine texts into a list
texts = [text1, text2]

tfidf_matrix = vectorizer.fit_transform(texts)

# -----------------------------------------
# Step 3: Compute Cosine Similarity
# -----------------------------------------

similarity_matrix = cosine_similarity(tfidf_matrix)

similarity_score = similarity_matrix[0][1]

# -----------------------------------------
# Step 4: Display Results
# -----------------------------------------

print("Text 1:", text1)
print("Text 2:", text2)
print("\nSimilarity Score:", round(similarity_score, 3))

# -----------------------------------------
# Step 5: Function for Custom Comparison
# -----------------------------------------

def check_similarity(t1, t2):
    texts = [t1, t2]
    tfidf = vectorizer.fit_transform(texts)
    sim = cosine_similarity(tfidf)[0][1]
    return sim

# Example custom comparison
new_text1 = "Deep learning is part of AI"
new_text2 = "Bananas are yellow fruits"

score = check_similarity(new_text1, new_text2)

print("\nCustom Comparison:")
print("Text A:", new_text1)
print("Text B:", new_text2)
print("Similarity Score:", round(score, 3))
```
