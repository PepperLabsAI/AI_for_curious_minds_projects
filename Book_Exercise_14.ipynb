```python
# ============================================
# Question Answering using BERT
# Beginner Friendly
# Ready for Google Colab
# ============================================

# Install required libraries (run once in Colab)
!pip install transformers torch

from transformers import pipeline

# -----------------------------------------
# Step 1: Load Pretrained BERT QA Model
# -----------------------------------------

qa_pipeline = pipeline("question-answering")

print("BERT Model Loaded Successfully!")

# -----------------------------------------
# Step 2: Provide Context (Paragraph)
# -----------------------------------------

context = """
BERT stands for Bidirectional Encoder Representations from Transformers.
It was developed by Google in 2018.
BERT is designed to understand the context of words in a sentence.
It has significantly improved performance in many NLP tasks such as question answering and sentiment analysis.
"""

# -----------------------------------------
# Step 3: Ask a Question
# -----------------------------------------

question = "Who developed BERT?"

# -----------------------------------------
# Step 4: Get Answer
# -----------------------------------------

result = qa_pipeline(question=question, context=context)

print("\nQuestion:", question)
print("Answer:", result["answer"])
print("Confidence Score:", round(result["score"], 3))

# -----------------------------------------
# Step 5: Try More Questions
# -----------------------------------------

question2 = "When was BERT developed?"

result2 = qa_pipeline(question=question2, context=context)

print("\nQuestion:", question2)
print("Answer:", result2["answer"])
print("Confidence Score:", round(result2["score"], 3))
```
