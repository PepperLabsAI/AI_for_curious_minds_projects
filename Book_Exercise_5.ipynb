🛍 Problem Statement

A retail store wants to group customers based on their purchasing behavior.

They collected two features:

Annual Spending (in thousands)

Number of Store Visits per Month

The store does not know the customer categories beforehand.

Your task:

Use K-Means Clustering to group customers into clusters.

Since this is unsupervised learning, there is no target variable.

```python
# ============================================
# Exercise 5: K-Means Clustering
# Customer Segmentation
# Beginner Friendly
# Ready for Google Colab
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# -----------------------------------------
# Step 1: Create Dataset
# -----------------------------------------

# Annual Spending (in thousands)
spending = [15, 18, 16, 20, 22, 80, 85, 78, 90, 88, 40, 45, 42, 38, 50]

# Store Visits per Month
visits = [2, 3, 2, 4, 3, 10, 12, 11, 13, 12, 6, 7, 6, 5, 8]

# Combine into feature matrix
X = np.array(list(zip(spending, visits)))

# -----------------------------------------
# Step 2: Apply K-Means
# -----------------------------------------

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("Cluster Centers:")
print(centroids)

# -----------------------------------------
# Step 3: Visualize Clusters
# -----------------------------------------

plt.scatter(X[:, 0], X[:, 1], c=labels)

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200)

plt.xlabel("Annual Spending (Thousands)")
plt.ylabel("Store Visits per Month")
plt.title("Customer Segmentation using K-Means")
plt.show()
```
