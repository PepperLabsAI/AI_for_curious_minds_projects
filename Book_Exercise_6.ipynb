🧬 Problem Statement

A fitness research center collected the following measurements of individuals:

Height (cm)

Weight (kg)

Chest Size (cm)

These three features are correlated.

The researcher wants to:

👉 Reduce 3 dimensions into 2 dimensions
👉 Visualize the data in simpler form
👉 Keep maximum important information

Your task:

Use PCA to reduce the dataset from 3 features to 2 principal components.

```python
# ============================================
# PCA - Dimensionality Reduction
# Beginner Friendly
# Ready for Google Colab
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# -----------------------------------------
# Step 1: Create Dataset
# -----------------------------------------

# Height (cm), Weight (kg), Chest Size (cm)
data = np.array([
    [170, 65, 90],
    [160, 55, 85],
    [180, 75, 95],
    [175, 70, 92],
    [165, 60, 88],
    [155, 50, 82],
    [185, 80, 100],
    [172, 68, 91],
    [168, 62, 89],
    [178, 72, 94]
])

print("Original Shape:", data.shape)

# -----------------------------------------
# Step 2: Standardize Data
# PCA works better when data is scaled
# -----------------------------------------

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# -----------------------------------------
# Step 3: Apply PCA
# Reduce from 3 dimensions to 2
# -----------------------------------------

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

print("Reduced Shape:", reduced_data.shape)

# -----------------------------------------
# Step 4: Print Explained Variance
# -----------------------------------------

print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)

# -----------------------------------------
# Step 5: Visualize Reduced Data
# -----------------------------------------

plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA - 2D Representation of 3D Data")
plt.show()
```
