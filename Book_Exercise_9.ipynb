📌 Problem Statement

In computer vision, edge detection helps identify object boundaries in images.

Your task:

Generate a simple image with shapes

Convert it to grayscale

Detect edges using Canny Edge Detection

Visualize the results

```python
# ============================================
# Image Edge Detection using OpenCV
# Beginner Friendly
# Ready for Google Colab
# ============================================

# Install OpenCV (only first time in Colab)
!pip install opencv-python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------
# Step 1: Create a Simple Image
# -----------------------------------------

# Create blank white image
image = np.ones((400, 400, 3), dtype=np.uint8) * 255

# Draw shapes
cv2.rectangle(image, (50, 50), (150, 150), (0, 0, 255), -1)   # Red square
cv2.circle(image, (300, 100), 50, (255, 0, 0), -1)            # Blue circle
cv2.line(image, (50, 300), (350, 300), (0, 255, 0), 5)        # Green line

# Convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# -----------------------------------------
# Step 2: Convert to Grayscale
# -----------------------------------------

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# -----------------------------------------
# Step 3: Apply Canny Edge Detection
# -----------------------------------------

edges = cv2.Canny(gray, 50, 150)

# -----------------------------------------
# Step 4: Display Results
# -----------------------------------------

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(edges, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")

plt.show()
```
