```python
# ============================================
# Stateful Rule-Based Chatbot
# Beginner Friendly Version
# Ready to Run in Google Colab
# ============================================

# We will create a chatbot that:
# 1. Remembers user's name
# 2. Remembers user's mood
# 3. Remembers user's favorite subject
# 4. Responds differently based on context

print("🤖 Hello! I am your friendly chatbot.")
print("Type 'exit' anytime to stop.\n")

# -----------------------------
# Step 1: Create Memory Storage
# -----------------------------

# We use a dictionary to store conversation memory
memory = {
    "name": None,
    "mood": None,
    "favorite_subject": None
}

# -----------------------------
# Step 2: Start Chat Loop
# -----------------------------

while True:

    user_input = input("You: ").lower()

    # Exit condition
    if user_input == "exit":
        print("Chatbot: Goodbye! Have a great day 😊")
        break


    # -----------------------------------------
    # Rule 1: User tells their name
    # -----------------------------------------
    elif "my name is" in user_input:
        name = user_input.replace("my name is", "").strip()
        memory["name"] = name
        print(f"Chatbot: Nice to meet you, {name.title()}!")


    # -----------------------------------------
    # Rule 2: User asks chatbot to recall name
    # -----------------------------------------
    elif "what is my name" in user_input:
        if memory["name"]:
            print(f"Chatbot: Your name is {memory['name'].title()} 😊")
        else:
            print("Chatbot: I don't know your name yet. Tell me by saying 'My name is ...'")


    # -----------------------------------------
    # Rule 3: User shares mood
    # -----------------------------------------
    elif "i am feeling" in user_input:
        mood = user_input.replace("i am feeling", "").strip()
        memory["mood"] = mood
        print(f"Chatbot: I understand. You are feeling {mood}.")


    # -----------------------------------------
    # Rule 4: Chatbot responds based on mood
    # -----------------------------------------
    elif "how am i feeling" in user_input:
        if memory["mood"]:
            print(f"Chatbot: You told me you are feeling {memory['mood']}.")
        else:
            print("Chatbot: You haven't told me your mood yet.")


    # -----------------------------------------
    # Rule 5: Favorite subject memory
    # -----------------------------------------
    elif "my favorite subject is" in user_input:
        subject = user_input.replace("my favorite subject is", "").strip()
        memory["favorite_subject"] = subject
        print(f"Chatbot: Wow! {subject.title()} is interesting.")


    elif "what is my favorite subject" in user_input:
        if memory["favorite_subject"]:
            print(f"Chatbot: Your favorite subject is {memory['favorite_subject'].title()}.")
        else:
            print("Chatbot: I don't know your favorite subject yet.")


    # -----------------------------------------
    # Rule 6: Greeting
    # -----------------------------------------
    elif "hello" in user_input or "hi" in user_input:
        if memory["name"]:
            print(f"Chatbot: Hello {memory['name'].title()}! 😊")
        else:
            print("Chatbot: Hello! What is your name?")


    # -----------------------------------------
    # Rule 7: Default Response
    # -----------------------------------------
    else:
        print("Chatbot: I am still learning. Can you say that in a different way?")
```
